```python
import pandas as pd

from nltk.corpus import inaugural as presidents

president = []
for fileid in presidents.fileids():
    tag, filename = fileid.split('-')
    president.append((filename, tag, presidents.raw(fileid)))

df = pd.DataFrame(president, columns=['President', 'year', 'text'])
```

```python
df.head(1)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
import matplotlib.pyplot as plt
import string
import re
```

```python
# Add stopwords
stop_words=stopwords.words('english')
add_to_stop_words=[',','.','--','?','(',')','-',';',"'",";",]
stop_words.extend(add_to_stop_words)
stop_words=set(stop_words)
```

```python
stop_words
```

```python
df.head(10)
```

```python
result= df.iloc[[38,43,46]]
```

```python
result.reset_index()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
result['Word_count']=result['text'].apply(lambda x: len(str(x).split(" ")))
result[['President','text','Word_count']].head()
```

```python
result['Stopwords']=result['text'].apply(lambda x: len([x for x in x.split() if x in stop_words]))
result[['text','Stopwords']].head()
```

# Basic Pre-Processing

Lower Case conversion

```python
result['text'] = result['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
result['text'].head()
```

```python
result['text']=result['text'].str.replace('[^\w\s]','')
```

```python
result['text'].head()
```

```python
# Removal of StopWords
result['text']=result['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop_words))
result['text'].head()
```

```python
result ['Word_count_Aftr_StopWords']=result['text'].apply(lambda x: len(str(x).split(" ")))
result[['text','Word_count_Aftr_StopWords']].head()
```
